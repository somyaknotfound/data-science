```python
class Atm:
  #constructor(special function) it can execute without being called explicitly
  def __init__(self):
    self.pin = ''
    self.balance = 0
    print("I am a constructor")
    self.menu()

  def menu(self):
    user_input = input("""
    Hey how can I help you?
    1. Press 1 to create pin
    2. Press 2 to change pin
    3. Press 3 to check balance
    4. Press 4 to withdraw
    5. Anything else to exit
    """)

    if user_input == '1':
      self.create_pin()
    elif user_input == '2':
      self.change_pin()
    elif user_input == '3':
      self.check_balance()

    elif user_input == '4':
      self.withdraw()
      pass
    else:
      exit()
  def create_pin(self):
    self.pin = input("Enter your pin")
    print("pin Set Successfully")

    user_balance = int(input("Enter Balance"))
    self.balance = user_balance
    self.menu()

    print("Pin created Successfully")
  def change_pin(self):
    old_pin = input("Enter old pin")

    if old_pin == self.pin:
      new_pin = int(input("Enter new pin"))
      self.pin = new_pin
      print("New pin created Successfully")
      self.menu()
    else:
      print("Invalid Pin")
      self.menu()

  def check_balance(self):
    curr_balance = self.balance
    print(f"Your balance is {curr_balance}")
    self.menu()

  def withdraw(self):
    user_balance = self.balance
    withdraw_amount = int(input("Enter withdraw amount"))
    if withdraw_amount < user_balance:
      self.balance = user_balance-withdraw_amount
      print("Withdraw Successfully")
    else:
      print("Insufficient Balance")
    self.menu()





```

```python
class fraction:

  def __init__(self,x,y):
    self.num = x
    self.den = y
  def __str__(self):
    return '{}/{}'.format(self.num,self.den)


  def __add__(self,other):
    new_num = other.den*self.num +other.num*self.den
    new_den = other.den*self.den
    return '{}/{}'.format(new_num,new_den)
```

```python
frac1 = fraction(3,4)
frac2 = fraction(1,2)
```

```python
print(frac1+frac2)
```
