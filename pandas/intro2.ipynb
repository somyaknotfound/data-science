```python
import pandas as pd
import numpy as np
```

```python
df_exams = pd.read_csv('StudentsPerformance.csv')
df_exams.head()
```

CREATE INDEX

```python
import random
```

```python
#create non repetetive values for the index
new_index = np.arange(0,1000)
```

```python
#shuffle
random.shuffle(new_index)
```

```python
new_index
```

```python
#create new column with new index
df_exams['new_index'] = new_index
```

```python
df_exams.head()
```

SET INDEX


```python
#set the new_index as the new index of the data frame
df_exams.set_index('new_index',inplace=True)
```

```python
df_exams.head()
```

SORT INDEX


```python
df_exams.sort_index(ascending=False)
```
