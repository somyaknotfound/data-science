adding col using assign and insert


use assign when you have to change the already existing values in a column

```python
import numpy as np
```

```python
#create random numbers
score1 = np.random.randint(1,100,size=1000)
score2 = np.random.randint(1,100,size=1000)
```

```python
#create a series using random numbers
series1 = pd.Series(score1 , index = np.arange(0,1000))
series2 = pd.Series(score2 , index = np.arange(0,1000))
```

```python
#using assign function
df_exams.assign(score1 = series1, score2 = series2)
```

```python
df_exams.head()
```

```python
#using index()
df_exams.insert(1,'test1',series1)
```
